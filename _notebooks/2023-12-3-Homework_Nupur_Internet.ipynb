#### Models
1. Three of the four following protocols are on the same layer. Identify which ones and what layer they are on, and why they are on each layer:
    - ASCII (see above for information):  Not a protocol but a character encoding standard. It's not associated with a particular layer in the OSI model.
    - FTP (facilitates transfer of files over the internet): Operates at the Application Layer (Layer 7) as it deals with file transfer functions.
    - TLS (see HTTPS section): Functions at the Transport Layer (Layer 4), providing secure communication between applications.
    - USB (permits data exchange between electronics): Not a network protocol; it's a physical interface that permits data exchange between electronic devices.


2. Telnet is a internet protocol which allows remote access to other computers over a local network or the internet. What layer of the OSI model would this protocol be located on? What is the function of this layer?
Telnet operates at the Application Layer (Layer 7) of the OSI model. This layer facilitates user interfaces and network applications.

2. TCP has error checking, which ensures that all packets arrive properly. Why is this important?
<br>
 - Give an example of how TCP ensures that there are no errors.

 TCP's error checking ensures data integrity during transmission by using acknowledgment and retransmission. For instance, if a packet is lost, TCP retransmits it until it's successfully received.
